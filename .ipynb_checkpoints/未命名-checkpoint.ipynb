```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from decord import cpu, AVReader, VideoReader, AudioReader, gpu
```

```python
def get_images_decord(fname):
    vr = VideoReader(fname, num_threads=0, ctx=cpu(0))
    seg_len = len(vr)
    sample_rate_scale = 1
    clip_len=8
    all_index = []
    # 随机选取中间一段
    converted_len = 8
    end_idx = np.random.randint(converted_len, seg_len)
    str_idx = end_idx - converted_len
    # 从start_idx到end_idx之间选取8个点，相邻点间距相等
    # 例如 2 18 8 -> [ 2.          4.28571429  6.57142857  8.85714286 11.14285714 13.42857143 15.71428571 18.        ]
    index = np.linspace(str_idx, end_idx, num=clip_len)
    # 去除小数位，向下取整，得到采样帧对应的索引
    index = np.clip(index, str_idx, end_idx - 1).astype(np.int64)
    index = index
    all_index.extend(list(index))

    all_index = all_index[::int(sample_rate_scale)]
    vr.seek(0)
    buffer = vr.get_batch(all_index).asnumpy()
    return buffer
```
